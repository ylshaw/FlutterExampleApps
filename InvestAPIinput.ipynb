```
import investpy
import pandas as pd
import matplotlib.pyplot as plt

ticker = input('Your stock code: ')

tickerDf = investpy.get_stock_historical_data(stock=ticker,
                                    country='malaysia',
                                    from_date='01/08/2019',
                                    to_date='05/08/2020')
tickerDf.head()
```

```
tickerDf = tickerDf['Close']

# Calculate the 20 and 100 days moving averages of the closing prices
short_rolling_tickerDf = tickerDf.rolling(window=20).mean()
long_rolling_tickerDf = tickerDf.rolling(window=50).mean()

fig, ax = plt.subplots(figsize=(16,9))

ax.plot(tickerDf.index, tickerDf, label=ticker)
ax.plot(short_rolling_tickerDf.index, short_rolling_tickerDf, label='20 days rolling')
ax.plot(long_rolling_tickerDf.index, long_rolling_tickerDf, label='50 days rolling')


ax.set_xlabel('Date')
ax.set_ylabel('Closing Price (RM)')
ax.legend();
```
